## Cells
Jupyter notebooks consist of different cells.  Cells can contain code, text markdown, or raw NBconvert.  

### Code Cell Example

```python
# This is an example of a code cell.
# Everything in this cell must be proper Python syntax.
# If you select this cell and hit "Run", the code below with run and the results will be displayed below this cell.

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(-2*np.pi,2*np.pi,100)

for i,c in enumerate(['orange','green','purple']):
    plt.plot(x,np.sin(i*x),color=c,label='A = {}'.format(i))
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = sin(Ax)')
plt.legend()
```

### Working with Cells

## Kernel
The kernel is the program that executes the code contained within your notebook.  Currently, you are running a Python 2 kernel.  You could attempt to run this notebook with a different kernel (say Python 3 or R), but it would probably produce a lot of errors because the code syntax is not correct.  You can manage the kernal with the Kernel tab above.

<img src="files/ScreenShots/kernel_menu.png">

- Interupt - This will stop whatever process is currently running.  This is useful if your code ends up in an infinite loop.
- Restart - This will restart the whole kernel session.  Note that restarting the kernel will remove all variables that have previously been assigned from memory, so you may have to rerun code cells.
- Restart & Clear Output - This will clear all output.
- Restart & Run All - This will restart the kernel and rerun all your cells from the begining.
- Change Kernel - This will change what kernel you're using.

Restarting the kernel WILL remove all stored variables from memory.  This means you may have to rerun some cells in your notebook.  Restarting the kernel will also show how your notebook will run next time you (or someone else) opens it, so it is a good idea to try running notebooks after restarting to make sure your code doesn't have dependencies on variables that were save in memory but deleted from your code.
