### Using ExternalStorage

This notebook shows you how to mount external cloud storage, such as Google Drive using a tool call ExternalStorage. This tool requires 2 things:

1) A directory where your cloud storages will be mounted, and

2) a directory where the credentials needed to authenticate the storages will be stored.

The first step is to import the tool:

```python
import ExternalStorage
```

To see the available storage options:

```python
ExternalStorage.print_available()
```

To mount storage from an available service, we use the Service class which has syntax:

Service(path_to_credentials_directory,path_to_mount_location,storage=shortname)

So to mount Google Drive, we provide the shortname: "google":

```python
google = ExternalStorage.Service('/path/to/creds','/path/to/mnt',storage='google')
```

From here we can run "google.authenticate()" and follow the instructions. You will need to copy paste a URL into a new browser window, log in to your Google account and allow the request. When you allow the request, Google will provide you with a code that you paste in to the text box below.

```python
google.authenticate()
```

Then we can mount google drive to the '/path/to/mnt/google' directory using:

```python
google.mount()
```

```python
import os
os.listdir('/path/to/mnt/google')
```

Then to unmount, we can just use:

```python
google.unmount()
```
